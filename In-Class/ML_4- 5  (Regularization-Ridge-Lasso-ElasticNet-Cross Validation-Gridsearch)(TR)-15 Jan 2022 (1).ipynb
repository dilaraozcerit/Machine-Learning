# Regularization(Ridge, Lasso, ElasticNet) - Cross Validation - Gridsearch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False)

poly_features = polynomial_converter.fit_transform(X)

#biz burada overfitting görebilmek için datadaki feature sayısını include_bias ile arttırıyorum. 
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

## Scaling the Data

```python
# standardization = z-score almak aslında
# normalization = mevcut sutunu en küçük dğerden çıkartıp max ve min değere bölüyoruz.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scaler = StandardScaler()

#minmaxscaler getdummi sonucu çıkan 1 0 larla daha çok tercih ediliyor
#robust olan da outlierlardan müzdarip olan datalar kullanılıyor

#data birbirine çok yakınsa zaten sclae yapmana gerek yoktur. 
#Eğer arada kaldıysan scale yapma konuusnda yap gitsin zararı yok


```

```python
scaler.fit(X_train)

#bilgi sızıntısını engellemek için yapılır
#hepsini z-score cinsine çevirdi. Bu yüzden bir kayma olabiliyor. 
#Buna engel olmam lazım.

#bu datada sadece x-traine ait bilgiler var.

# X traine ait ortalama vs std bilgilerini alıyor.
# burdaki fit eğitim yapmıyor. Burada bilgileri topluyor.
# fit varsa sadece traini yazıyoruz.
# x test i predict in içine koyuyoruz.
```

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled

#yukarıdaki ilk denklemi yapıyor. Ortalamyı çıkarıp std a bölüyor. Z-score'a çevirdi. 

# şimdi eğitim başlıyor. Her biri yeni Z-score lardan oluşan yeni bir datamız oluştu.
                                            # bilgileri x trainden alıyoruz sadece.
                                            # x teste dair hiçbir bilgi yok elimizde. data x testten hiçbir şey öğrenmedi.
                                            # scale işlemini yaparken fit işlemini sadece x traine uyguluyoruz.
                                            # burdaki dönüşüm x train üzerinden alınan verilerle yapılıyor. 
                                            # data leakage'i engelliyoruz biz burada.

# Her dataya ortalama vs std ile işlem yapıldığı için diğerlerine ait verileri de alıyor bu da >> data leakage (test ve trainin aynı anda fitlenmesi sonucu oluşur)

#X-test : sınav soruları (soruları görürse kopya olur, veri sızıntısı var demektir.)
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

### checking std = 1 and mean = 0

this gives us the z-scores. so it's also called z-score 

These values show where in the normal distribution they correspond to the z score.


```python
pd.DataFrame(X_train_scaled).agg(["mean", "std"]).round()

#mean 0 std 1 olduğunu buradan teyit etmiş olduk 
```

```python
pd.DataFrame(X_test_scaled).agg(["mean", "std"]).round()

#z-score a bakmamın sebebi etki oranının farklı olması yüzünden train ve test de bakıyorum
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, y_train)

#artık işim sclae edilmiş data ile. Targeta kesinlikle scale uygulamıyorum. 
#Çünkü onlarla test ediyorum zaten modeli
```

```python
y_pred = lm.predict(X_test_scaled)
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls

#variance(r2) değerlerine baktığımızda uçurum var. Demek ki ezber yapmış
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)

#Multicollinearity exists whenever an independent variable is highly correlated with one or more of the other independent variables in a multiple regression equation.
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV


# avantajı : en iyi skoru döndürecek hyperparametreleri veriyor
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42)

#alpha cezalandırmyı güçlendirmek için vardı. Ya da iyi çıkan sonuçları biraz olsun kötüleştirip gerçekliğe yakınlaştırıyorum 

# burada özellikle ElasticNet kullandık
# l1_ratio her zaman lasso dur. l2 de ridge'dir.
# l1_Ratio = Lasso nun ne kadar kullanılacağı
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}

#hypermetre sayınız ne kadar fazla olursa o kadar iterasyon: tekerrür/tekrarlama demek. Bu da güçsüz bilgisayar ile zor bir iş. 
#iki parametrenin aynı anda kesişen noktalarını belirlemeye çalışacağım. Dict şeklinde yazmak lazım. 

#iki tane alpha parametresei var. Özellikle 0a çekmiyoruz. Zaten uyarı alırsınız. Eğer öyle yaparsanız zaten linear reg. e dönüşür model. 

# ne kadar çok hyperparametre o kadar çok işlem zamanı
# deneye yanıla modelimizdeki en iyi parametleri bulmaya çalışacağız. sınır değerler çıkarsa tekrar deneyeceğiz.
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)

#hangi model üzerine grid uygulayacaksak, gelip o modele bakacağım. 
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_

#en düşük skoru kim veriyorsa onları belirledi. Lassonun aynısını verdi bana.
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
grid_model.best_index_

#en iyi değerin indexini söyledi. 
```

```python
grid_model.best_score_

#scorelar arasındaki en iyi scoreları bu veriyor.
```

Hiper parametrelerden bazıları sonsuz sayıda değer alabilecek konumdadırlar. Bununla birlikte biz problem hakkında sahip olduğumuz ön bilgileri 
kullanarak hiper parametrelerin alabilecekleri değerler için aralıklar belirleyebiliriz. Belirlediğimiz bu aralıklardan belirli ana noktalar 
seçilerek hiper parametreler için değer listeleri oluşturulur. Grid search ile hiper parametre seçim işleminde; belirlenen aralıkta bulunan tüm 
değerlerin kombinasyonları için ağ eğitilip sonuçlar gözlenir duruma göre en iyi kombinasyon hiper parametre grubu olarak seçilir.
