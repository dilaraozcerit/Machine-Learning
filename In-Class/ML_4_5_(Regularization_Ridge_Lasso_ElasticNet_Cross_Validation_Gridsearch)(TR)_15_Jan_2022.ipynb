```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures

# polynomial yapyoruz nk overfittingi grmek iin sonra da ridge ve lasso ile nasl dzelteceiz ona bakacaz
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False) # include_bias derken her birine 1 ekler
# degree ?????????
poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)  # x'i atmadk atsaydk linear olurdur, onun yerine poly_features attk
```

## Scaling the Data

```python
# scale yapyoruz ki nemi abartlmasn.

# standardization = z-score almak aslnda
# normalization = mevcut sutunu en kk derden kartp max ve min deere blyoruz.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler     

scaler = StandardScaler()
```

```python
scaler.fit(X_train)  # X traine ait ortalama vs std bilgilerini alyor.
                     # burdaki fit eitim yapmyor. Burada bilgileri topluyor.
                     # fit varsa sadece traini yazyoruz.
                     # x test i predict in iine koyuyoruz.
```

```python
X
# Her dataya ortalama vs std ile ilem yapld iin dierlerine ait verileri de alyor bu da >> data leakage
```

```python
X_train_scaled = scaler.transform(X_train)  # imdi eitim balyor. Her biri yeni Z-score lardan oluan yeni bir datamz olutu.
X_train_scaled                              # bilgileri x trainden alyoruz sadece.
                                            # x teste dair hibir bilgi yok elimizde. data x testten hibir ey renmedi.
                                            # scale ilemini yaparken fit ilemini sadece x traine uyguluyoruz.
                                            # burdaki dnm x train zerinden alnan verilerle yaplyor. 
                                            # data leakage'i engelliyoruz biz burada.
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled

# hibir test datas ile fit yapmyoruz, aksi takdirde data leakage oluur.
```

### checking std = 1 and mean = 0

this gives us the z-scores. so it's also called z-score 

These values show where in the normal distribution they correspond to the z score.


```python
pd.DataFrame(X_train_scaled).agg(["mean", "std"]).round()    # round diyoruz nk zaten ou 0'a yakn deerler
```

```python
pd.DataFrame(X_test_scaled).agg(["mean", "std"]).round()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

# target deerlerine yani y-test'e scale uygulamyoruz
```

```python
lm.fit(X_train_scaled, y_train) # y'yi scale yapmyoruz.
```

```python
y_pred = lm.predict(X_test_scaled)  # aralarnda overfitting var m yok mu diye bunlar alyoruz.
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls          # train ile test arasnda variance uurumu var
            # overfit olduk. ( 3 kat 10 kat fark var) regularization lazm.
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
pd.DataFrame(poly_features) # bunlar dataframe deerlerimiz
```

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'     # grselletirme iin kodumuz
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)  # imdi korelasyon vakti B)  # multicollinearityleri grdk
                                                              # multicol, feature selection'e engel oluyor
                                                              # multicol, linear modellerde sknt, advanced modellerde bu absorb edilebiliyor.
                                                              
```

cross validation is only applied to the train set.

## Grid Search

```python
from sklearn.model_selection import GridSearchCV # avantaj : en iyi skoru dndrecek hyperparametreleri veriyor 
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42) # burada zellikle ElasticNet kullandk
# l1_ratio her zaman lasso dur. l2 de ridge'dir.
# l1_Ratio = Lasso nun ne kadar kullanlaca
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}

# ne kadar ok hyperparametre o kadar ok ilem zaman
# deneye yanla modelimizdeki en iyi parametleri bulmaya alacaz. snr deerler karsa tekrar deneyeceiz. 
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)

# param girid
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_
```
