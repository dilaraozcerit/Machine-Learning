```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures

# polynomial yapıyoruz çünkü overfittingi görmek için sonra da ridge ve lasso ile nasıl düzelteceğiz ona bakacağız
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False) # include_bias derken her birine 1 ekler
# degree ?????????
poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)  # x'i atmadık atsaydık linear olurdur, onun yerine poly_features attık
```

## Scaling the Data

```python
# scale yapıyoruz ki önemi abartılmasın.

# standardization = z-score almak aslında
# normalization = mevcut sutunu en küçük dğerden çıkartıp max ve min değere bölüyoruz.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler     

scaler = StandardScaler()
```

```python
scaler.fit(X_train)  # X traine ait ortalama vs std bilgilerini alıyor.
                     # burdaki fit eğitim yapmıyor. Burada bilgileri topluyor.
                     # fit varsa sadece traini yazıyoruz.
                     # x test i predict in içine koyuyoruz.
```

```python
X
# Her dataya ortalama vs std ile işlem yapıldığı için diğerlerine ait verileri de alıyor bu da >> data leakage
```

```python
X_train_scaled = scaler.transform(X_train)  # şimdi eğitim başlıyor. Her biri yeni Z-score lardan oluşan yeni bir datamız oluştu.
X_train_scaled                              # bilgileri x trainden alıyoruz sadece.
                                            # x teste dair hiçbir bilgi yok elimizde. data x testten hiçbir şey öğrenmedi.
                                            # scale işlemini yaparken fit işlemini sadece x traine uyguluyoruz.
                                            # burdaki dönüşüm x train üzerinden alınan verilerle yapılıyor. 
                                            # data leakage'i engelliyoruz biz burada.
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled

# hiçbir test datası ile fit yapmıyoruz, aksi takdirde data leakage oluşur.
```

### checking std = 1 and mean = 0

this gives us the z-scores. so it's also called z-score 

These values show where in the normal distribution they correspond to the z score.


```python
pd.DataFrame(X_train_scaled).agg(["mean", "std"]).round()    # round diyoruz çünkü zaten çoğu 0'a yakın değerler
```

```python
pd.DataFrame(X_test_scaled).agg(["mean", "std"]).round()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

# target değerlerine yani y-test'e scale uygulamıyoruz
```

```python
lm.fit(X_train_scaled, y_train) # y'yi scale yapmıyoruz.
```

```python
y_pred = lm.predict(X_test_scaled)  # aralarında overfitting var mı yok mu diye bunları alıyoruz.
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls          # train ile test arasında variance uçurumu var
            # overfit olduk. ( 3 kat 10 kat farkı var) regularization lazım.
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
pd.DataFrame(poly_features) # bunlar dataframe değerlerimiz
```

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'     # görselleştirme için kodumuz
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)  # şimdi korelasyon vakti B)  # multicollinearityleri gördük
                                                              # multicol, feature selection'e engel oluyor
                                                              # multicol, linear modellerde sıkıntı, advanced modellerde bu absorb edilebiliyor.
                                                              
```

We do cross-validation to check whether the one-time scores we receive are consistent or not

cross validation is only applied to the train set.

## Grid Search

```python
from sklearn.model_selection import GridSearchCV # avantajı : en iyi skoru döndürecek hyperparametreleri veriyor 
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42) # burada özellikle ElasticNet kullandık
# l1_ratio her zaman lasso dur. l2 de ridge'dir.
# l1_Ratio = Lasso nun ne kadar kullanılacağı
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}

# ne kadar çok hyperparametre o kadar çok işlem zamanı
# deneye yanıla modelimizdeki en iyi parametleri bulmaya çalışacağız. sınır değerler çıkarsa tekrar deneyeceğiz. 
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)

# param girid
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_
```

## Using Best Hyper Parameters From GridSearch

```python
y_pred = grid_model.predict(X_test_scaled)
y_train_pred = grid_model.predict(X_train_scaled)
```

```python
train_val(y_train, y_train_pred, y_test, y_pred, "GridSearch")
```
