## Train test split

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X=df.drop("Selling_Price", axis=1)
y=df.Selling_Price
```

__Encoder__

Modelimiz tree based bir model oldugu icin dummy yerine encoder islemini uygulayacagiz :

```python
cat = X.select_dtypes("object").columns
cat
```

```python
X[cat].head()
```

```python
enc = OrdinalEncoder()
X[cat] = enc.fit_transform(X[cat])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
print("Train features shape : ", X_train.shape)
print("Train target shape   : ", y_train.shape)
print("Test features shape  : ", X_test.shape)
print("Test target shape    : ", y_test.shape)
```

## Modeling for Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
DT_model = DecisionTreeRegressor(random_state=101)
```

```python
DT_model.fit(X_train, y_train)
```

Modelimizi egittikten sonra default degerler ile skorlarimiza bakalim :

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {"train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    "test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
train_val(DT_model, X_train, y_train, X_test, y_test)
```

Yukarida rmse degerleri birbirinden uzak cikmis, yani overfitting durumu var. Ama asagidadaha once  get_dummiesle alidigimiz sonuclar ile kiyaslarsak biraz iyilesme oldugunu goruyoruz. Demek ki tree based modellerde Encoder uygulamanin model uzerine olumlu etkisi var. 

!!!!! Regression analizlerinde bir modelin "pure" olmasi cok zordur (Tree model olsa bile). O yuzden train test skorlari birbirine yakin olsa bile direk overfitting var diyebiliriz. Classification' da "pure" skorlar gozlenebilir.  !!!!!

__Cross Validate__

Cross Validate skorlari da tek seferlik skorlarla uyumlu.

```python
from sklearn.model_selection import cross_validate, cross_val_score
model = DecisionTreeRegressor(random_state=101)
scores = cross_validate(model, X_train, y_train, scoring=['r2', 
            'neg_mean_absolute_error','neg_mean_squared_error','neg_root_mean_squared_error'], cv =10)
df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]
```

Yukaridaki denklemde 'model.estimators_[0]' icindeki indexi her degistirdigimizde baska bir agaca gecer. Classification' da her agacta Root Node' un farkli farkli oldugunu goruyorduk ama burda Regression analizinde hep ayni feature' in Root Node oldugunu goruyoruz. Bunun sebebi nedir?  Burdaki 'max_feature' parametresinin de default='auto'. Yani classification ile ayni. Classification' da feature sayisinin karekokunu kadar random feature secip aralarindan en iyisini seciyordu. Ama burda butun feature' lari her seferinde kullanir. Bu acidan Tree Base' den farki yok, tek farki boostrap yapmasi (Yani her agacta datanin 2/3' unu alir.)

!!! Random Forest Regression analizlerinde GridSearch yaparken "max_feature" parametresini mutlaka bulundur ve default degerde birakma. !!!

Arastirmalara gore bu degeri de( feature sayisi / 3) olarak kullanmaliyiz. 

## GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestRegressor(random_state=101)
```

```python
param_grid = {"n_estimators":[64, 128, 200, 500],     
              "max_depth": [None, 4, 5, 6, 7,8],
              "min_samples_leaf": [1, 2, 3, 4],
              "min_samples_split": [2, 3, 5, 6],
              "max_features":['auto', len(df.columns)/3, 2, 4]}   

# n_estimators=64, 128 mutlaka denenmeli.
# Yukarida belirttigimiz gibi max_features : len(df.columns) / 3' u de deniyoruz. 
```

```python

grid_model = GridSearchCV(estimator=model,
                          param_grid=param_grid,
                          scoring='neg_root_mean_squared_error',
                          cv=10,
                          n_jobs = -1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_score =pd.DataFrame(grid_model.cv_results_)
grid_score
```

GridSearch islemi sonucu train ve test seti skorlari birbirine yaklasti.
