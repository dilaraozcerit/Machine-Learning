## PCA

```python
from sklearn.decomposition import PCA # boyut azaltma
```

```python
pca = PCA(n_components=2) 
# en önemli hyperparametremiz 'n_components'dir.
# bir de sonuçları daha tutarlı almak için random_state de verilebilir.
# Datamızdaki feature ve gözlem sayısına bakar, hangisi minimum ise onu alır.
# Gözlem sayısı min ise gözlem sayısı kadar, feature sayısı min ise feature sayısı kadar component alır.
```

 fit dediğimizde arka planda her yeni oluşan 2 kompanentin öz vektörlerini (eigen vectors) featurelarımız ile çarpıyor ve elde ettiği değerler ile yeni  bileşenler oluşturuyor.

```python
principal_components = pca.fit_transform(scaled_X)
```

```python
principal_components
# 30 feature'um vardı, şimdi bunları konsantre hale getirdi.
```

```python
component_df = pd.DataFrame(data = principal_components, columns = ["first_component", 
                                                                    "second_component"])
component_df
# daha okunaklı olması için df'e attık.
```

## Eigenvalues & Eigenvectors

*Bütün komponentlerde bütün feature lar var.
oran gibi değerlendirme, hangisi büyükse katkısı o kadar fazla diye düşün.*

* Oluşturacağım yeni componentlere (bileşenlere) feature ların ne kadar katkısı olacağını gösterir.
bu vector ve value covariance matrix üzerinden linear transformasyon ile elde ediliyor.



* Bunlar + veya - olabilirler. Eğer ikisi de + ise benim oluşturduğum yeni component ile (bileşenim ile) featuremın pozitif linear ilişkide olduğunu görüyorum

aynı cins malzemeden iki çorba ama malzemelerin kullanım oranı birbirinden farklı olduğu için birbirinden iki ayrı çorba

```python
#Eigenvectors
# dönüşümleri kullanırken eigenvectors kullanıyoruz.
pca.components_
# ilk listem 1. componenti kullanırken kullanacağı öz vektörler,
# ikinci listem 2. componenti kullanırken kullanacağı öz vektörlerim.
```

```python
eigenvectors_first_component = pca.components_[0]
eigenvectors_first_component
# bu birincisiydi.
```



 genelde ilk componentlerde featurelardan daha çok katkı vardır.
mesela ilk componentte 40% katkı daha sonraki 25% daha sonrakinde 10% .... gibi ilerler.

 bizim datamızda şuan 30 tane feature var. Fakat burdaki bilgileri nasıl aktarıyoruz? >> Eigenvectors ile.

```python
first_eigenvectors = pd.DataFrame(eigenvectors_first_component, index=X.columns, 
                                  columns=["first_eigenvectors"]).sort_values("first_eigenvectors", ascending=False)

first_eigenvectors
# bunları sıralayarak gösteriyoruz.
# ilk feature'un ne kadar katkısı olduğunu gösteriyor. ilki mesela 0.26. ama bunlar oran değil, sadece sayısal büyüklük.
# her bileşende her feature'un mutlaka katkısı var. Az belki ama var.
# Artık ben hangi feature'un ne kadar katkı sağladığını biliyorsam, componentlerimi oluşturabilirim.
```

```python
eigenvectors_second_component = pca.components_[1]
eigenvectors_second_component
```

İlk komponentte katkısı fazla olan feature mın ikincisinde katkı katsayısı azalıyor. Benim bütün bileşenlerimde bütün feature larımın katkısı var. Ama çok ama az. Ama şunu biliyorum. Bir bileşende katkısı fazla ise diğerlerinde azalıyor. Yani featurelarım genelde bir komponentte toplanıyor.

Şimdi manuel olarak yapıyoruz yukardakinin aynısını.

```python
second_eigenvectors = pd.DataFrame(eigenvectors_second_component, index=X.columns, 
                                   columns=["second_eigenvectors"]).sort_values("second_eigenvectors", ascending=False)

second_eigenvectors
# 2. componentimi oluşturmak için kullancağım özvektör değerlerlerim.
```

```python
pd.concat([first_eigenvectors, second_eigenvectors], axis=1)
# bazıları mesela ilkinde çok etki etmiş ama ikincide çok etmiş.
# bu iki component neye karar veriyor?
# aklımıza şöyle bir soru geliyor >> aacaba bir component kanseri teşhis etmede kullanılıyor olabilir mi?
# bir tanesi kanser olanları bir tanesi kanser olmayanları anlatıyor olabilir. ama bu sadece benim fikrim şimdilik.

# ama bazılarında (bkz. mean compactness) neredeyse eşit derecede etki ediyorlar. Acaba bazı featurelarım da kanseri teşhiş etmede çok önemli değil mi?
# Eşeğin aklına karpuz kabuğunu düşürmüşüz burada.
```

```python
df_scaled_X.loc[0] # scale edilmiş datamdaki ilk değerime ait bütün featurelarımı aldım.
```

```python
(df_scaled_X.loc[0] * eigenvectors_first_component).sum() # ilk değerimi first componentim ile çarpıyorum 
```

```python
(df_scaled_X.loc[0] * eigenvectors_second_component).sum()  # yine aynı değerimi bu sefer ikinci componentimin ile çarpıyorum.
```

## Finding optimal number of components

```python
pca.explained_variance_ratio_  
# Percentage of variance explained by each of the selected components.
# Elimde olan toplam bilginin 44%'ünü 1. componentte, 2. componentte 10%'unu tutuyorum.
```

```python
pca.explained_variance_ratio_.cumsum()
# burada kümülatif toplamlarını yaptık: çıkan sonuç >> biz 2 feature ile var olan datamızın 63%'ünü elimizde tutuyoruz. 
```

```python
pca = PCA().fit(scaled_X)
x = range(1,31)
plt.plot(x, np.cumsum(pca.explained_variance_ratio_),)
plt.xlabel("Component count")
plt.ylabel("Variance Ratio")
# kümülatif toplamlarını görselleştiriyoruz burada.
```

Tavsiye edilen %75 - %95. Datamız Linear ve güçlü corr var ise ama ek bilgi olarak, özellikle datam lineare uygunsa ve güçlü featurelarım varsa 75% lik orana da takılmamak lazım.

## K-Means Algorithm

değerin 0 - ile 0.5 arasında olması gerekir. Ne kadar 0 a yakınsa kümelemeye o kadar iyidir.

```python
hopkins(component_df, component_df.shape[0])
```

```python
from sklearn.cluster import KMeans

ssd = []

K = range(2,10) 
# bunu neden 1'den başlatmıyoruz? çünkü 1'den 2'ye düşerken inertia değeri çok keskin bir düşüş sergileyeceği için hep 2 ve 3'e meyleder.

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_df)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")

# elbow metodumuz bize 3 demiş
```

```python
ssd
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
# keskin düşüşün durduğu nokta 3.
```

```python
df_diff.plot(kind='bar')
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(component_df)        # Fit the data to the visualizer
visualizer.show();

# ama yellowbrick 4 demiş...
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_df)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_df, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
# burdan şunu anlıyoruz. 2 de seçsek 3 de seçsek yanlış bir karar olmaz. ikisi de olabilir.
# yani datam şöyle olabilir ya kanser / kanser değil. ya da 2. , 3. ya da 4. evre kanser datalarının olduğu bir data olabilir.
# bu çıkarımı neye göre yaptık? >> tüm skorlarımız birbirine çok yakın.
```

```python
sns.scatterplot(x = component_df.first_component, y= component_df.second_component, alpha=0.7)
# solda çok güzel kümelenme var ama sağ ve orta kısımlar geniş yayılmış. 2 mi 3 mü emin değiliz.
```

```python
model = KMeans(n_clusters =2, random_state=42)
clusters = model.fit_predict(component_df)
clusters
```

```python
ct = pd.crosstab(y, clusters)
ct
```
