## PCA

```python
from sklearn.decomposition import PCA # boyut azaltma
```

```python
pca = PCA(n_components=2) 
# en nemli hyperparametremiz 'n_components'dir.
# bir de sonular daha tutarl almak iin random_state de verilebilir.
# Datamzdaki feature ve gzlem saysna bakar, hangisi minimum ise onu alr.
# Gzlem says min ise gzlem says kadar, feature says min ise feature says kadar component alr.
```

 fit dediimizde arka planda her yeni oluan 2 kompanentin z vektrlerini (eigen vectors) featurelarmz ile arpyor ve elde ettii deerler ile yeni  bileenler oluturuyor.

```python
principal_components = pca.fit_transform(scaled_X)
```

```python
principal_components
# 30 feature'um vard, imdi bunlar konsantre hale getirdi.
```

```python
component_df = pd.DataFrame(data = principal_components, columns = ["first_component", 
                                                                    "second_component"])
component_df
# daha okunakl olmas iin df'e attk.
```

## Eigenvalues & Eigenvectors

*Btn komponentlerde btn feature lar var.
oran gibi deerlendirme, hangisi bykse katks o kadar fazla diye dn.*

* Oluturacam yeni componentlere (bileenlere) feature larn ne kadar katks olacan gsterir.
bu vector ve value covariance matrix zerinden linear transformasyon ile elde ediliyor.



* Bunlar + veya - olabilirler. Eer ikisi de + ise benim oluturduum yeni component ile (bileenim ile) featuremn pozitif linear ilikide olduunu gryorum

ayn cins malzemeden iki orba ama malzemelerin kullanm oran birbirinden farkl olduu iin birbirinden iki ayr orba

```python
#Eigenvectors
# dnmleri kullanrken eigenvectors kullanyoruz.
pca.components_
# ilk listem 1. componenti kullanrken kullanaca z vektrler,
# ikinci listem 2. componenti kullanrken kullanaca z vektrlerim.
```

```python
eigenvectors_first_component = pca.components_[0]
eigenvectors_first_component
# bu birincisiydi.
```



 genelde ilk componentlerde featurelardan daha ok katk vardr.
mesela ilk componentte 40% katk daha sonraki 25% daha sonrakinde 10% .... gibi ilerler.

 bizim datamzda uan 30 tane feature var. Fakat burdaki bilgileri nasl aktaryoruz? >> Eigenvectors ile.

```python
first_eigenvectors = pd.DataFrame(eigenvectors_first_component, index=X.columns, 
                                  columns=["first_eigenvectors"]).sort_values("first_eigenvectors", ascending=False)

first_eigenvectors
# bunlar sralayarak gsteriyoruz.
# ilk feature'un ne kadar katks olduunu gsteriyor. ilki mesela 0.26. ama bunlar oran deil, sadece saysal byklk.
# her bileende her feature'un mutlaka katks var. Az belki ama var.
# Artk ben hangi feature'un ne kadar katk saladn biliyorsam, componentlerimi oluturabilirim.
```

```python
eigenvectors_second_component = pca.components_[1]
eigenvectors_second_component
```

lk komponentte katks fazla olan feature mn ikincisinde katk katsays azalyor. Benim btn bileenlerimde btn feature larmn katks var. Ama ok ama az. Ama unu biliyorum. Bir bileende katks fazla ise dierlerinde azalyor. Yani featurelarm genelde bir komponentte toplanyor.

## K-Means Algorithm

deerin 0 - ile 0.5 arasnda olmas gerekir. Ne kadar 0 a yaknsa kmelemeye o kadar iyidir.

```python
pip install pyclustertend
```

```python
from pyclustertend import hopkins
```

```python
hopkins(component_df, component_df.shape[0])
```

```python
from sklearn.cluster import KMeans

ssd = []

K = range(2,10) 
# bunu neden 1'den balatmyoruz? nk 1'den 2'ye derken inertia deeri ok keskin bir d sergileyecei iin hep 2 ve 3'e meyleder.

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_df)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")

# elbow metodumuz bize 3 demi
```

```python
ssd
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
# keskin dn durduu nokta 3.
```

```python
df_diff.plot(kind='bar')
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(component_df)        # Fit the data to the visualizer
visualizer.show();

# ama yellowbrick 4 demi...
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_df)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_df, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
# burdan unu anlyoruz. 2 de sesek 3 de sesek yanl bir karar olmaz. ikisi de olabilir.
# yani datam yle olabilir ya kanser / kanser deil. ya da 2. , 3. ya da 4. evre kanser datalarnn olduu bir data olabilir.
# bu karm neye gre yaptk? >> tm skorlarmz birbirine ok yakn.
```

```python
sns.scatterplot(x = component_df.first_component, y= component_df.second_component, alpha=0.7)
# solda ok gzel kmelenme var ama sa ve orta ksmlar geni yaylm. 2 mi 3 m emin deiliz.
```

```python
model = KMeans(n_clusters =2, random_state=42)
clusters = model.fit_predict(component_df)
clusters
```

```python
ct = pd.crosstab(y, clusters)
ct
```

53 hata yapm ve neredeyse ounu bilmi.
