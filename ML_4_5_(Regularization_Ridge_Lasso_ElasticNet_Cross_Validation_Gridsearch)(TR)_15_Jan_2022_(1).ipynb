```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls = train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

Bağımsız değişkenler arasında kuvvetli correlasyon varsa...

Bunu önlemezsem modelim için çok önemli olan bir future ı kaçırabilirim. çünkü correlasyondan dolayı normalde önemli olan future ımı düşük etkili gösterebilir.def color_red(val):
    
    if val > 0.90 and val < 1.0:
        color = 'red'
    elif val == 1.0:
        color = 'orange'
    else:
        color = 'black'
    return f'color: {color}'

Bağımsız değişkenler arasında kuvvetli correlasyon varsa...

Bunu önlemezsem modelim için çok önemli olan bir future ı kaçırabilirim. çünkü correlasyondan dolayı normalde önemli olan future ımı düşük etkili gösterebilir.

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'

# %90 ve üzeri bir multicorolasyon göstergesidir.
```

R2 neden eksi? Burada eğitim geröekleşmemiş yani bir tutarsızlık ve overfitting durumu vardır. Eğitim namına bir şey gerçekleşmemiş.

- R2 [R2r2] compares the fit of the chosen model with that of a horizontal straight line (the null hypothesis). 
    - ***[Seçilen modelin uyumunu yatay bir düz çizgininkiyle (boş hipotez) karşılaştırır.]***
- If the chosen model fits worse than a horizontal line, then  R2 is negative. 
    - ***[Seçilen model yatay bir çizgiden daha kötü uyuyorsa, R2R2 negatiftir.]*** 
- Note that R2 is not always the square of anything, so it can have a negative value without violating any rules of math.
    - ***[R2R2'nin her zaman hiçbir şeyin karesi olmadığını unutmayın, bu nedenle herhangi bir matematik kuralını ihlal etmeden negatif bir değere sahip olabilir.]*** 
- R2 is negative only when the chosen model does not follow the trend of the data, so fits worse than a horizontal line. 
    - ***[R2R2, yalnızca seçilen model verilerin trendini takip etmediğinde negatiftir, bu nedenle yatay bir çizgiden daha kötü uyuyor.]***

## Grid Search

Bundan sonra hyper-parametrelerimiz olacak. Bunların hangisi iyi skor veriyor deneme yanılma yapma şansımız yok.
GridSearchCV ile hangi modelle çalışacaksak estimate olarak onu giriyoruz.  param_grid olarak  hangi alpha ve hangi l1_ratio aralıklarında aramasını istiyorsam onu tanımlıyorum ki onların kombinasyonunu alıp en iyi alfa ve en iyi l1_raito'yu versin. skoring ile de hangi error skorunu çalışacağını belirliyoruz. burada root_mean_squared_error kullanmasını istedik.

Ayrıca crossvalidation yapmanıza gerek yok çünkü GridSearch onu bizim için yapıyor.

ridge ve lasso sadece alpha yı kullanır. (alpha ya ilave olarak)  l1_ratio yu elastic te kullanıyoruz. 

l1_ratio değeri ile lassoya ne oranda benzediğini anlıyoruz; 1 ise %100 lasso oluyor.

Yani Grid Search; benim parametrelerimin olabilecek tüm kombinasyonuna 10'ar defa crossValidation yapıyor ve yine parametre ile belirlediğimiz error skoruyla çalışan bir model oluşturuyor.

lassoCV, lasso için geçerlidir. ridgecv ridge için geçerlidir, elasticnet, elasticnet ile geçerlidir. ama ben GridSearch ile içine hangisinin parametresini koyarsam onunla çalışır. dolayısıyla tüm modellere uygulanabilir.

```python
from sklearn.model_selection import GridSearchCV

# avantajı : en iyi skoru döndürecek hyperparametreleri veriyor
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42)

# burada özellikle ElasticNet kullandık
# l1_ratio her zaman lasso dur. l2 de ridge'dir.
# l1_Ratio = Lasso nun ne kadar kullanılacağı
```
