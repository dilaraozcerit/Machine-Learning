# Regularization(Ridge, Lasso, ElasticNet) - Cross Validation - Gridsearch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures

# polynomial yapÄ±yoruz Ã§Ã¼nkÃ¼ overfittingi gÃ¶rmek iÃ§in sonra da ridge ve lasso ile nasÄ±l dÃ¼zelteceÄiz ona bakacaÄÄ±z
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False) # include_bias derken her birine 1 ekler
# degree ?????????
poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)  # x'i atmadÄ±k atsaydÄ±k linear olurdur, onun yerine poly_features attÄ±k
```

## Scaling the Data

```python
# scale yapÄ±yoruz ki Ã¶nemi abartÄ±lmasÄ±n.

# standardization = z-score almak aslÄ±nda
# normalization = mevcut sutunu en kÃ¼Ã§Ã¼k dÄerden Ã§Ä±kartÄ±p max ve min deÄere bÃ¶lÃ¼yoruz.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler     

scaler = StandardScaler()
```

```python
scaler.fit(X_train)  # X traine ait ortalama vs std bilgilerini alÄ±yor.
                     # burdaki fit eÄitim yapmÄ±yor. Burada bilgileri topluyor.
                     # fit varsa sadece traini yazÄ±yoruz.
                     # x test i predict in iÃ§ine koyuyoruz.
```

```python
X
# Her dataya ortalama vs std ile iÅlem yapÄ±ldÄ±ÄÄ± iÃ§in diÄerlerine ait verileri de alÄ±yor bu da >> data leakage
```

```python
X_train_scaled = scaler.transform(X_train)  # Åimdi eÄitim baÅlÄ±yor. Her biri yeni Z-score lardan oluÅan yeni bir datamÄ±z oluÅtu.
X_train_scaled                              # bilgileri x trainden alÄ±yoruz sadece.
                                            # x teste dair hiÃ§bir bilgi yok elimizde. data x testten hiÃ§bir Åey Ã¶Ärenmedi.
                                            # scale iÅlemini yaparken fit iÅlemini sadece x traine uyguluyoruz.
                                            # burdaki dÃ¶nÃ¼ÅÃ¼m x train Ã¼zerinden alÄ±nan verilerle yapÄ±lÄ±yor. 
                                            # data leakage'i engelliyoruz biz burada.
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled

# hiÃ§bir test datasÄ± ile fit yapmÄ±yoruz, aksi takdirde data leakage oluÅur.
```

### checking std = 1 and mean = 0

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

# target deÄerlerine yani y-test'e scale uygulamÄ±yoruz
```

```python
lm.fit(X_train_scaled, y_train) # y'yi scale yapmÄ±yoruz.
```

```python
y_pred = lm.predict(X_test_scaled)  # aralarÄ±nda overfitting var mÄ± yok mu diye bunlarÄ± alÄ±yoruz.
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls          # train ile test arasÄ±nda variance uÃ§urumu var
            # overfit olduk. ( 3 kat 10 kat farkÄ± var) regularization lazÄ±m.
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
pd.DataFrame(poly_features) # bunlar dataframe deÄerlerimiz
```

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'     # gÃ¶rselleÅtirme iÃ§in kodumuz
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)  # Åimdi korelasyon vakti B)  # multicollinearityleri gÃ¶rdÃ¼k
                                                              # multicol, feature selection'e engel oluyor
                                                              # multicol, linear modellerde sÄ±kÄ±ntÄ±, advanced modellerde bu absorb edilebiliyor.
                                                              
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV # avantajÄ± : en iyi skoru dÃ¶ndÃ¼recek hyperparametreleri veriyor 
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42) # burada Ã¶zellikle ElasticNet kullandÄ±k
# l1_ratio her zaman lasso dur. l2 de ridge'dir.
# l1_Ratio = Lasso nun ne kadar kullanÄ±lacaÄÄ±
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}

# ne kadar Ã§ok hyperparametre o kadar Ã§ok iÅlem zamanÄ±
# deneye yanÄ±la modelimizdeki en iyi parametleri bulmaya Ã§alÄ±ÅacaÄÄ±z. sÄ±nÄ±r deÄerler Ã§Ä±karsa tekrar deneyeceÄiz. 
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)

# param girid
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```
